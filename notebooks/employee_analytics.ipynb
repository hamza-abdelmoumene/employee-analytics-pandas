## Part 1 — NumPy Fundamentals

#### 1.1 Basic Statistics

```python
salary_array = df['salary'].dropna().to_numpy()

print("Mean:  ", np.mean(salary_array))
print("Median:", np.median(salary_array))
print("Std:   ", np.std(salary_array))
print("Min:   ", np.min(salary_array))
print("Max:   ", np.max(salary_array))
```

#### 1.2 Normalization

```python
normalized_salary_array = (salary_array - salary_array.min())/(salary_array.max() - salary_array.min())

print(normalized_salary_array[:5])  # should be values between 0 and 1, that's what a normalized array means!
print(f"Min: {normalized_salary_array.min()}, Max: {normalized_salary_array.max()}")
```

#### 1.3 Salary Level Classification

```python
salary_level = np.where(salary_array < 80000, "Low",
               np.where(salary_array <= 110000, "Mid", "High"))
# Or:
'''conditions = [
    salary_array < 80000,
    salary_array <= 110000
]
choices = ["Low", "Mid"]
salary_level = np.select(conditions, choices, default="High")'''
print(salary_level[:10])
print(np.unique(salary_level, return_counts=True))  # shows how many Low/Mid/High
```

#### 1.4 Percentiles

```python
# Calculate percentiles
p25 = np.percentile(salary_array, 25)   # Q1 (25th)
p50 = np.percentile(salary_array, 50)   # Median (50th)  
p75 = np.percentile(salary_array, 75)   # Q3 (75th)
p90 = np.percentile(salary_array, 90)   # 90th

print(f"25th percentile (Q1):  ${p25:,.0f}")
print(f"50th percentile (Median): ${p50:,.0f}")
print(f"75th percentile (Q3):  ${p75:,.0f}")
print(f"90th percentile:       ${p90:,.0f}")
```

#### 1.5 2D Array Operations

```python
salary_simulation = np.random.randint(50000, 150001, size= (5, 3)) # simulating 5 employees × 3 months of salary.
mean_salary_per_emp = salary_simulation.mean(axis= 1)
mean_salary_per_mnth = salary_simulation.mean(axis= 0)

print(salary_simulation)
print(f"Mean salary per employee: {mean_salary_per_emp}")
print(f"Mean salary per month: {mean_salary_per_mnth}")
```

## Part 3 — Data Cleaning

#### 3.1 Column Names

```python
# Standardize all column names: lowercase, strip whitespace, replace spaces with underscores:

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns)
```

#### 3.2 Smart Salary Fill

```python
# Fill missing salary values with the median salary of that employee's department — not the global median:

df['salary'] = df['salary'].fillna(
    df.groupby('department')['salary'].transform('median')
)

print(df['salary'])
```

#### 3.3 Satisfaction Score Fill

```python
# Fill missing satisfaction_score with the column mean. Round to 1 decimal:

df['satisfaction_score'] = df['satisfaction_score'].fillna(df['satisfaction_score'].mean().round(1))
print(df['satisfaction_score'])
```

#### 3.4 Unknown Location:

```python
# Fill missing city and country with the string "Unknown":

df[['city', 'country']] = df[['city', 'country']].fillna('Unknown')
print(df[['city', 'country']])
```

#### 3.5 Promotion Date Fill

```python
# Fill missing last_promotion_date with the employee's own hire_date:

df['last_promotion_date'] = df['last_promotion_date'].fillna(df['hire_date'])
print(df['last_promotion_date'])
```

#### 3.6 Boolean Conversion

```python
# Make sure is_manager and is_dept_head are proper boolean dtype:

if (df['is_dept_head'].dtypes != bool):
    df['is_dept_head'] = df['is_dept_head'].astype(bool)
    print("is_dept_head type successfully set as boolean")
else:
    print("is_dept_head type is already boolean")
    
if (df['is_manager'].dtypes != bool):
    df['is_manager'] = df['is_manager'].map({True: True, False: False, 'True': True, 'False': False}).fillna(False)
    print("is_manager type successfully set as boolean")
else:
    print("is_manager type is already boolean")
```

#### 3.7 Drop Bad Rows

```python
# Remove rows where both salary AND bonus are missing at the same time:

print(f"Before Removing: \ndf size = {df.size}")

df = df.dropna(subset= ['salary', 'bonus'], how= 'all')

print(f"After Removing: \ndf size = {df.size}")
df.head()
```

```python
print("Remaining NaNs:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print(f"\nShape after cleaning: {df.shape}")
```

## Part 4 — Feature Engineering:

#### 4.1 Total Compensation

```python
# Create a total_compensation column = salary + bonus:

total_compensation = df['salary'] + df['bonus']
df['total_compensation'] = df['salary'] + df['bonus']
print(total_compensation)
```

#### 4.2 Bonus Percentage

```python
# Create bonus_pct = bonus as a percentage of salary. Round to 2 decimals: 

bonus_pct = (df['bonus'] / df['salary']) * 100
bonus_pct = bonus_pct.round(2)
df['bonus_pct'] = (df['bonus'] / df['salary'] * 100).round(2)
print(bonus_pct)
```

#### 4.3 Years Since Promotion

```python
# Create years_since_promotion = number of years between last_promotion_date and today.

years_since_promotion = (pd.Timestamp.now() - df['last_promotion_date']).dt.days
years_since_promotion /= 365.25
df['years_since_promotion'] = years_since_promotion
print(years_since_promotion.round(2))
```

#### 4.4 Tenure

```python
# Create tenure_years = how many years the employee has been with the company (from hire_date to today):

tenure_years = (pd.Timestamp.now() - df['hire_date']).dt.days
tenure_years /= 365.25
df['tenure_years'] = tenure_years
print(tenure_years.round(2))
```

#### 4.5 Seniority Labels

```python
# Create a seniority column using pd.cut() on tenure_years:

df['seniority'] = pd.cut(tenure_years, 
                         bins=[0, 2, 5, 10, np.inf], 
                         labels=['Junior', 'Mid', 'Senior', 'Lead']) 
print(df['seniority'])
```

#### 4.6 Salary Band

```python
# Create salary_band using pd.qcut() on salary into 4 equal groups labeled Q1, Q2, Q3, Q4:

df['salary_band'] = pd.qcut(df['salary'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
df['salary_band'].value_counts()
```

#### 4.7 Remote Flag

```python
# Create a boolean column is_remote = True if remote_days_per_week >= 3: 

df['is_remote'] = df['remote_days_per_week'] >= 3
print(df['is_remote'])
```

```python
# From hire_date, extract three new columns: hire_year, hire_month, hire_day_of_week (as name e.g. "Monday"):

df['hire_year'] = df['hire_date'].dt.year
df['hire_month'] = df['hire_date'].dt.month
df['hire_day_of_week'] = df['hire_date'].dt.day_name()
print(df[['hire_date', 'hire_year', 'hire_month', 'hire_day_of_week']].head())
```

```python
df = df = df.drop(columns='hire_day_of_week')   
df.head()
```

## Part 6 — GroupBy & Aggregation

#### 6.1 Department Summary

```python
# For each department compute: headcount, mean salary, max salary, min salary, median bonus.
# Sort by mean salary descending: 

grouped_dept = df.groupby('department').agg(
    headcount=('employee_id', 'count'),
    mean_salary=('salary', 'mean'),
    max_salary=('salary', 'max'),
    min_salary=('salary', 'min'),
    median_bonus=('bonus', 'median')
).sort_values(by='mean_salary', ascending=False)

grouped_dept.head(len(grouped_dept))
```

#### 6.2 Status by Country

```python
# For each country; 
# compute the percentage of active vs non-active employees:

country_status = df.groupby('country')['status'].value_counts(normalize=True) * 100
country_status.round(2)
print(country_status)
```

#### 6.3 Performance by Contract & Seniority

```python
# For each contract_type + seniority combination, compute average performance_rating. 
# Then unstack the result into a readable table:

perfo_by_cont_senio = df.groupby(['contract_type', 'seniority'], observed=False)['performance_rating'].mean().unstack()
print(perfo_by_cont_senio)
```

#### 6.4 Salary vs Department Average

```python
# Use .transform() to add dept_avg_salary = mean salary of each employee's department.
# Then add salary_vs_dept_avg = difference between employee salary and their dept average:

df['dept_avg_salary'] = df.groupby('department')['salary'].transform('mean')
df['salary_vs_dept_avg'] = df['salary'] - df['dept_avg_salary']

print(df[['first_name', 'department', 'salary', 'dept_avg_salary', 'salary_vs_dept_avg']].head(10))
```

#### 6.5 Filter Low Satisfaction Departments

```python
# Keep only employees from departments where the average satisfaction score is above 5: 

filter_employees = df.groupby('department').filter(lambda group: group['satisfaction_score'].mean() > 5)

print(filter_employees.shape)
print("\nDepartments kept:")
print(filter_employees['department'].value_counts())
```

#### 6.6 Named Aggregations

```python
# Using the modern .agg() named syntax, create a clean department summary with columns: headcount, avg_salary, avg_bonus, avg_performance, avg_satisfaction:

departments_summary = df.groupby('department').agg(
                      headcount=('employee_id', 'count'),
                      avg_salary=('salary', 'mean'),
                      avg_bonus=('bonus', 'mean'),
                      avg_performance=('performance_rating', 'mean'),
                      avg_satisfaction=('satisfaction_score', 'mean')
                      ).round(2)
print(departments_summary)
```

#### 7.3 Cumulative Salary Spend

```python
# Sort by hire_date and compute the cumulative total salary as if you hired employees one by one over time:

sorted_hd = df.sort_values(by= 'hire_date') # hd stands for hire_date
sorted_hd['cumulative_salary'] = sorted_hd['salary'].cumsum()
print(sorted_hd[['hire_date', 'first_name', 'salary', 'cumulative_salary']].head(10))
```

```python
# After sorting by hire_date, compute a 3-employee rolling average of salary. 
# What trend do you notice? 

rolling_averrage = sorted_hd['salary'].rolling(window=3).mean()
print(rolling_averrage)
```

### Observation
The 3-employee rolling average is highly volatile because the window 
is too small. No clear trend is visible with window=3.

A larger window (e.g. 30 or 50) would smooth the curve and reveal 
whether salaries increased or decreased over time.

## Part 11 — Final Business Analysis

#### 11.1 Attrition by Department

```python
# Which department has the highest attrition rate? (attrition = Resigned or Terminated)


df['attrited'] = df['status'].isin(['Resigned', 'Terminated'])
print(f"Total attrited employees: {df['attrited'].sum()}")
dept_attrition = (df.groupby('department')
                  .agg({
                      'attrited': 'mean',        # Proportion
                      'employee_id': 'count'     # Headcount
                  })
                  .round(3)
                  .rename(columns={
                      'attrited': 'attrition_rate',
                      'employee_id': 'headcount'
                  })
)
dept_attrition['attrition_rate'] *= 100
dept_attrition = dept_attrition.sort_values('attrition_rate', ascending=False)
print("\nDepartment Attrition Summary:")
print(dept_attrition.head(7))
print(f"\nHighest attrition department: {dept_attrition.index[0]}")
```

#### 11.2 Experience vs Salary

```python
# Is there a correlation between years_experience and salary?

correlation = df['salary'].corr(df['years_experience'])
print(correlation)
```

### Observation
There's a moderately strong positive relationship between salary and experience years. As one increases, the other tends to increase too, but it's not a perfect relationship! 



#### 11.3 Managers vs Non-Managers

```python
# Do managers earn significantly more than non-managers on average? :

managers = df.loc[df['is_manager'], 'salary'].mean()
employees = df.loc[~df['is_manager'], 'salary'].mean()

print(f"Managers:  ${managers:,.0f}")
print(f"Employees: ${employees:,.0f}") 
print(f"Difference:${managers - employees:+,.0f}")
```

#### 11.4 — Education vs Performance 

```python


education = df.groupby('job_education')['performance_rating'].mean().sort_values(ascending=False)
print(education)
```

#### 11.5 Promotion Speed

```python
# What is the average time between hire_date and last_promotion_date in years? Which department promotes fastest?:

df['promotion_speed'] = (pd.to_datetime(df['last_promotion_date']) - 
                        pd.to_datetime(df['hire_date'])).dt.days / 365.25

dept_promotion = (df.groupby('department')['promotion_speed']
                  .mean()
                  .sort_values()  
                  .round(1)
)

print("Average years to first promotion by department:")
print(dept_promotion)
print(f"\n FASTEST PROMOTING: {dept_promotion.index[0]} ({dept_promotion.iloc[0]:.1f} years)")
```

##### 11.6 CEO Dashboard

```python
# Create a single clean summary DataFrame — one row per department
# containing: headcount, avg_salary, attrition_rate, avg_performance, avg_satisfaction, avg_tenure.
# Export it to ../outputs/dept_summary.csv.

ceo_dashboard = df.groupby('department').agg(
    headcount        = ('employee_id',        'count'),
    avg_salary       = ('salary',             'mean'),
    attrition_rate   = ('attrited',           'mean'),
    avg_performance  = ('performance_rating', 'mean'),
    avg_satisfaction = ('satisfaction_score', 'mean'),
    avg_tenure       = ('tenure_years',       'mean'),
).round(2)

ceo_dashboard['attrition_rate'] = (ceo_dashboard['attrition_rate'] * 100).round(1)
ceo_dashboard = ceo_dashboard.sort_values('avg_salary', ascending=False)

print(ceo_dashboard)

ceo_dashboard.to_csv('../outputs/dept_summary.csv')
```
